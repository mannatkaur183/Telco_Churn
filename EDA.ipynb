```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df=pd.read_csv("Telco-Customer-Churn.csv")
```

```python
df. head()
```

```python
df.shape
```

```python
df.info()
```

```python
df["TotalCharges"]=df["TotalCharges"].replace(" ","0")
df["TotalCharges"]=df["TotalCharges"].astype("float")
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df["customerID"].duplicated().sum()
```

```python
df["SeniorCitizen"]=df["SeniorCitizen"].map({0:'No',1:'Yes'})
```

#converted 0 and 1 values of senior citizen to yes /no to make it easier to understand

```python
df.head()
```

```python
ax=sns.countplot(x='Churn',data=df)
plt.title("Count of Customers by churn ")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
x=(df['Churn'].value_counts())
plt.pie(x,labels=x.index,autopct="%1.2F%%")
plt.title("Percentage of Churned Customers")
plt.show()
```

```python
# from the pie chart we can conclude that 26.54% of our customers have churned 
# lets explore more
```

```python
ax=sns.countplot(x="gender",data=df,hue='Churn',palette='dark:r')
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.show()
```

```python
ax=sns.countplot(x="SeniorCitizen",data=df,hue='Churn',palette='dark:r')
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.show()
```

```python
x=df.groupby('SeniorCitizen')['Churn'].value_counts(normalize=True).unstack()*100

ax=x.plot(kind='bar', stacked=True)
for bar in ax.patches:
    height = bar.get_height()
    if height > 0:
        ax.text(bar.get_x() + bar.get_width()/2,
                bar.get_y() + height/2,
                f'{height:.1f}%', ha='center', va='center', fontsize=8)

plt.ylabel('Percentage')
plt.legend(loc='upper right')
plt.show()
```

```python
# greater percentage of people in senior citizen category have churned
```

```python
plt.figure(figsize=(9,4))
sns.histplot(x='tenure',data=df,hue='Churn',bins=72)
plt.show()
```

```python
# people who used the services for a long time have stayed and others who used our services for 1 or 2 months have churned
```

```python
ax=sns.countplot(x='Contract',data=df,hue='Churn')
plt.title("Count of Customers by Contract")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
# people who have month to month are likely to churn then those who have one or two years of contract
```
